```python
grouped = startups.groupby(['City', 'State'])
print(grouped)
NumOfYes =0
NumOfNO=0
for (city, state), group in grouped:
    # Filter the group based on 'New Expansion' value if needed
    group = group[group['New Expansion'] == 'Old']
    if not group.empty:
        marketing_spend = group['Marketing Spend'].values[0]
        profit = group['Profit'].values[0]
        roms = group['ROMS'].values[0]
        
        print(f"City: {city}, State: {state}")
        
        if profit > 40053 and roms > 1274:
            NumOfYes = NumOfYes+1
            print(f"The Model suggests opening a new expansion in {city}, {state}")
        else:
            NumOfNO=NumOfNO+1
            print(f"The Model does not suggest opening a new expansion in {city}, {state}")
        
    else:
        print(f"No data found for City: {city}, State: {state}")
    
print(f"the number of suggestion for opening is  {NumOfYes}")
print(f"the number of suggestion for NOT opening is  {NumOfNO}")

```
