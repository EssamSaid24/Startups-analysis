```python
# Input City and State
input_city = 'Killeen'
input_state = 'Texas'

# Filter the DataFrame based on City and State
filtered_data = startups[(startups['City'] == input_city) & (startups['State'] == input_state) &(startups['New Expansion'] == 'Old')]

# Check if there's any matching data
if not filtered_data.empty:
    # Extract Marketing Spend, Profit, and ROMS values
    marketing_spend = filtered_data['Marketing Spend'].values[0]
    profit = filtered_data['Profit'].values[0]
    roms = filtered_data['ROMS'].values[0]
    print(f"Marketing Spend: {marketing_spend}")
    print(f"Profit: {profit}")
    print(f"ROMS: {roms}")

    if profit > 40053 and roms > 1274:
        print(f"The Model  suggests opening a new expansion in {input_city} city and {input_state} state ")
    else:
        print(f"The Model does not suggest opening a new expansion in {input_city} city and {input_state} state ")

else:
    print("Enter The Data Right")
```
